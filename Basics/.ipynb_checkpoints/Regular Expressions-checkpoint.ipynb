### Regular Expressions ( primary use: pattern recognition)
- finding the pattern of entire line

```python
import re

f = open('myfile')
for line in f:
    line= line.rstrip()
    a = input('enter a word: ')
    if re.search(a,line):          # re.search(pattern to be searched, where to search that pattern)'
        print(line)
f.close()
```

```python
import re

f = open('myfile')
for line in f:
    line= line.rstrip()
    if re.search('^kumar',line):          # re.search(^(startswith)pattern to be searched, where to search that pattern)'
        print(line)
f.close()
```

```python
# program to print line starting with a word in the form of Ixxm
import re

f = open('myfile')
for line in f:
    line= line.rstrip()
    if re.search('^I..m',line):          #print the lines that starts with 'I' in between 2 characters and 'm'
        print(line)
f.close()
```

```python
import re    # import regex --> regular expression

f = open('myfile')
for line in f:
    line= line.rstrip()
    if re.search('^k.+d',line):          # re.search(pattern to be searched, where to search that pattern)'
        print(line)
f.close()
```

```python
# write a program to extract lines from mbox-short.txt that starts with from and ends with .edu
import re

f = open('C:/Users/Administrator/Desktop/python/Python/mbox-short.txt','r')
count =0
for line in f:
    count+=1
    line= line.rstrip()
    if re.search('^[Ff]rom.*edu$',line):          # re.search(pattern to be searched, where to search that pattern)'
        print('linenumber:',count,'line:',line)
f.close()
```

```python
# Line should start with capital letters and ends with any digit
import re

f = open('C:/Users/Administrator/Desktop/python/Python/mbox-short.txt','r')
count =0
for line in f:
    count+=1
    line= line.rstrip()
    if re.search('^[A-Z].*[0-9]$',line):        # starts with capital letters and ends with any digit 
        print('linenumber:',count,'line:',line)
f.close()
```

```python
# Findall() funtion --> searches specified word with given pattern
# program to extract only domain name from email id using findall() and find() functions 
# (\S--> not whitespace , + --> matches 1 or more occurrence of domain name preceeding @ ,
# \S+ --> non whitesapce and matches the occurence preceeding expression)

import re

s = 'krishna, you have a message from masalaprui@gmail.edu.net and klasm@gmail.edu.net about job offer @4PM'
ls =re.findall('\S+@\S+',s)  
print(ls)
for email in ls:
    x = email.find('@')
    print(email[x+1:])
```

```python
import re

s= open('C:/Users/Administrator/Desktop/python/Python/mbox-short.txt','r') 
ls=[]
pattern ='[a-zA-Z0-9]\S+@\S+[a-zA-Z][.][a-zA-Z]+'
for line in s:
    line= line.strip()
    x = re.findall(pattern, line)
    if len(x)>0:
        ls.extend(x)
ls
```
